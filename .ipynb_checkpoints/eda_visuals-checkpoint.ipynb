```python
# Load dataset
import pandas as pd
%matplotlib inline

winequality_df = pd.read_csv('winequality_edited.csv')
winequality_df.info()
```

```python
winequality_df['fixed_acidity'].plot(kind = 'hist')
```

```python
winequality_df['total_sulfur_dioxide'].plot(kind = 'hist')
```

```python
winequality_df['pH'].plot(kind = 'hist')
```

```python
winequality_df['alcohol'].plot(kind = 'hist')
```

#### from the histograms plotted, fixed acidity and alcohol are skewed to the right
